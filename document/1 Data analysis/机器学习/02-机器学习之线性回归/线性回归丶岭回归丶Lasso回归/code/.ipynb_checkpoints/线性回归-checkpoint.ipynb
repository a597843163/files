## 一、普通线性回归

### 1、原理

分类的目标变量是标称型数据，而回归将会对连续型的数据做出预测。

应当怎样从一大堆数据里求出回归方程呢？

假定输人数据存放在矩阵X中，而回归系数存放在向量W中。那么对于给定的数据X1, 预测结果将会通过

Y=X*W

给出。现在的问题是，手里有一些X和对应的Y,怎样才能找到W呢？

一个常用的方法就是找出使误差最小的W。这里的误差是指预测Y值和真实Y值之间的差值，使用该误差的简单累加将使得正差值和负差值相互抵消，所以我
们采用平方误差。

## 二、岭回归

### 1、原理

缩减系数来“理解”数据

如果数据的特征比样本点还多应该怎么办？是否还可以使用线性回归和之前的方法来做预测？

答案是否定的，即不能再使用前面介绍的方法。这是因为输入数据的矩阵X不是满秩矩阵。非满秩矩阵在求逆时会出现问题。

为了解决这个问题，统计学家引入了岭回归（ridge regression)的概念

缩减方法可以去掉不重要的参数，因此能更好地理解数据。此外，与简单的线性回归相比，缩减法能取得更好的预测效果。

岭回归是加了二阶正则项的最小二乘，主要适用于过拟合严重或各变量之间存在多重共线性的时候，岭回归是有bias的，这里的bias是为了让variance更小。

```python
#统计时候，统计身高，三列，三个属性，sg1，sg2，sg3，身高，身长，身宽
#共线性
#去重
```

#### 归纳总结

1.岭回归可以解决特征数量比样本量多的问题

2.岭回归作为一种缩减算法可以判断哪些特征重要或者不重要，有点类似于降维的效果

3.缩减算法可以看作是对一个模型增加偏差的同时减少方差

岭回归用于处理下面两类问题：

1.数据点少于变量个数

2.变量间存在共线性（最小二乘回归得到的系数不稳定，方差很大）

### 2、实例

岭回归一般用在样本值不够的时候

```python
# 4个属性   2个样本值
X_train = np.array([[2,1,2,1],[2,3,2,3]])
y_train = np.array([1.2,3])
```

使用普通线性回归

```python
lrg = LinearRegression()
lrg.fit(X_train,y_train)
```

```python
lrg.predict([[3,4,4,3]])
```

使用岭回归

```python
# 导岭回归包
from sklearn.linear_model import Ridge
```

```python
# Ridge 有一个alpha的值,如果alpha=0,则是正常的普通线性回归
ridge = Ridge()
ridge.fit(X_train,y_train)
```

```python
# 预测
ridge.predict([[3,4,4,3]])
```

```python
#  ridge.coef_的值是线性回归的解  .显示出来其中有两个数被抛弃
ridge.coef_
```

```python
# 将数据抛弃,再次验算
X_train = np.array([[1,1],[3,3]])
y_train = np.array([1.2,3])
```

```python
# 重新训练
ridge.fit(X_train,y_train)
lrg.fit(X_train,y_train)
```

```python
# 线性回归重新预测
lrg.predict([[4,3]])
```

```python
#  岭回归重新预测
ridge.predict([[4,3]])
```

深入研究岭回归

```python
X_train = np.arange(0,10)
X_train
```

```python
# 增加一列
X_train = X_train[:,np.newaxis]
X_train
```

```python
# numpy 广播机制
X_train = 1/(X_train + np.arange(1,11))
X_train
```

```python
# 训练数据
X_train.shape
```

```python
# 训练数据
y_train = np.ones(10)
y_train
```

```python
# 岭回归中alpha
# linspace 线性 的  log返回更小的值或者更大的值
alpha = np.logspace(-10,-2,200)
alpha
```

```python
# 大的值
np.logspace(1,10,10)
```

```python
# 生成岭回归实例  
# fit_intercept 截距
ridge = Ridge(fit_intercept=False)
```

```python
# 观察不同的alpha对ridge的影响
coefs = []
for a in alpha:
    ridge.set_params(alpha=a)
    ridge.fit(X_train,y_train)
    coefs.append(ridge.coef_)
```

```python
coefs
```

```python
# axes.get_xlim()
```

```python
# 绘制图形,观察不同的alpha对我们数据的影响是非常巨大的
plt.figure(figsize=(12,9))
# 生成画板
axes = plt.gca()

axes.plot(alpha,coefs)

# 调整画板以log的方式进行展示
axes.set_xscale('log')

# 把坐标设置成相反
axes.set_xlim(axes.get_xlim()[::-1])

# 设置提示
axes.set_xlabel('alpha')

axes.set_ylabel('weight/coef')

# 设置标题  regularization(正则)
# axes.title("alpha 的影响")
# axes.title('Ridge coefficients as a function of the regularization')
```

总结:

    岭回归中coef值越接近0,说明此数据不那么重要
   
    alpha值,要取不那么剧烈的地方

### 1、原理

```python
# 导包
from sklearn.linear_model import Lasso
```

```python
y
```

```python
X_train = x[:-20]
x_test = x[-20:]
Y_train = y[:-20]
y_test = y[-20:]
```

```python
from sklearn.model_selection import train_test_split
X_btrain,x_btest,Y_btrain,y_btest = train_test_split(x,y,test_size=0.1)
```

## 四、普通线性回归、岭回归与lasso回归比较

导包，导入sklearn.metrics.r2_score用于给模型打分

```python
sample,feture = 50, 200

# 生成随机数
x = np.random.randn(sample,feture)

# 系数
coef = 3*np.random.randn(feture)

# 将190个设置为0
index = np.arange(feture)

# 打乱index的顺序
np.random.shuffle(index)

# 系数:190个设置为0,留10个有意义的
coef[index[:-10]] = 0

# 生成训练数据y
y = np.dot(x,coef)

# 添加噪声
y += 0.01*np.random.normal(sample)
```

```python
display(x.shape,y.shape)
```

```python
coef
```

使用numpy创建数据X，创建系数，对系数进行处理，对部分系数进行归零化操作，然后根据系数进行矩阵操作求得目标值  
增加噪声

训练数据和测试数据

```python
sample//2
```

```python
X_train,Y_train = x[:sample//2],y[:sample//2]
x_test,y_test = x[sample//2:],y[sample//2:]
```

分别使用线性回归，岭回归，Lasso回归进行数据预测

```python
# 线性回归
# 生成线性回归实例
lrg = LinearRegression()
# 训练数据
lrg.fit(X_train,Y_train)
# 预测数据
y_lrg = lrg.predict(x_test)
```

```python
# 岭回归
ridge = Ridge()
ridge.fit(X_train,Y_train)
y_ridge = ridge.predict(x_test)
```

```python
# Lasso回归
lasso = Lasso()
lasso.fit(X_train,Y_train)
y_lasso = lasso.predict(x_test)
```

```python
y_ridge
```

```python
y_lasso
```

```python
y_lrg
```

数据视图，此处获取各个算法的训练数据的coef_:系数

```python
plt.figure(figsize=(12,9))
plt.subplot(221)
# 真实的coef值
plt.plot(coef,c='r',label='orignal')
plt.legend()

# lrg正常线性回归
plt.subplot(222)
plt.plot(lrg.coef_,c='b',label='linear regressor')
plt.ylim(ymin=-6,ymax=4.8)
plt.legend()

# 领回归
plt.subplot(223)
plt.plot(ridge.coef_,c='g',label='ridge regressor')
plt.ylim(ymin=-6,ymax=4.8)
plt.legend()

# Lasso回归
plt.subplot(224)
plt.plot(lasso.coef_,c='purple',label='Lasso')
plt.ylim(ymin=-6,ymax=4.8)
plt.legend()
```

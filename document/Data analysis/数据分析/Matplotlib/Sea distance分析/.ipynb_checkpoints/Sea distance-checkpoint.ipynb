各城市与海洋距离，最高温度，最低温度，最高湿度，最低湿度

```python
ferrara['temp'].max()
```

```python
# 各城市距离海边的距离
dists = []
# 最高温度
temp_max = []
# 最低温度
temp_min = []
# 最高湿度
hum_max = []
# 最低湿度
hum_min = []
for c in citys:
    dists.append(c['dist'].iloc[0])
    temp_max.append(c['temp'].max())
    temp_min.append(c['temp'].min())
    hum_max.append(c['humidity'].max())
    hum_min.append(c['humidity'].min())
```

```python
display(dists,temp_max)
```

显示最高温度与离海远近的关系

```python
plt.scatter(dists,temp_max)
```

观察发现，离海近的可以形成一条直线，离海远的也能形成一条直线。

首先使用numpy：把列表转换为numpy数组，用于后续计算。

分别以100公里和50公里为分界点，划分为离海近和离海远的两组数据

```python
# 先转换成numpy才能使用一些方法
x = np.array(dists)
y = np.array(temp_max)
```

```python
# 100公里之后的数据
# x1 表示距离小于100公里的海滨城市
x1 = x[x<100].reshape(5,1)
y1 = y[x<100].reshape(5,1)
display(x1,y1)
```

```python
# 50公里以后的数据
x2 = x[x>50].reshape(6,1)
y2 = y[x>50].reshape(6,1)
display(x2,y2)
```

使用支持向量机计算回归参数

```python
# 机器学习的模型
from sklearn.svm import SVR
```

```python
# 创建算法
# kernel 设置什么方式来处理
svr1 = SVR(kernel='linear')
svr2 = SVR(kernel='linear')

# 将数据交给算法,并且训练
svr1.fit(x1,y1)
svr2.fit(x2,y2)

# 预测数据
x_test1 = np.linspace(0,100,20).reshape(20,1)
x_test2 = np.linspace(50,350,20).reshape(20,1)
display(x_test1.shape,x_test2.shape)
```

```python
# 预测温度
y1_ = svr1.predict(x_test1)
y2_ = svr2.predict(x_test2)
display(y1_,y2_)
```

```python
# 画图显示数据
# 原来的点
plt.scatter(dists,temp_max)

# 绘制回归图
plt.plot(x_test1,y1_,'r')
plt.plot(x_test2,y2_,'g')
```

由于风向是360度，我们可以考虑使用玫瑰图（极坐标条形图）

首先自定义一个画图函数

```python
def show_rose(values,title):
    # 玫瑰图花瓣的个数 8,每一瓣45度
    n = 8
    # 设置角度  总共360度   每一瓣360/8度
    angle = np.arange(0,2*np.pi,2*np.pi/n)
    
    # 绘制的数据values   每个数据以半径来显示
    radius = np.array(values)
    # axis  代表x轴,y轴
    # axes   代表的是整个画面,画布
    # polar  == True 是否使用极坐标
    plt.axes([0,0,2,2],polar = True)
    
    # 设置颜色
    color = np.random.random(size = 24).reshape((8,3))
    
    plt.bar(angle, radius, color = color)
    plt.title(title,loc = 'left')
```

```python
x = np.array([1,1.2,1,2,3.6,3,4])
# x数据  
# 4  分成几份
# 怎么分,从1,4进行等分
#返回第一个array是数据数量,统计的数据
# 第二个array是分割标准(距离)
np.histogram(x,4,[1,4])
```

```python
degree = milano['wind_deg']
d,b = np.histogram(degree,8,[0,360])
show_rose(d,'milano')
```

用numpy创建一个直方图，将360度划分为8个面元，将数据分类到这8个面元中

计算米兰各个方向上的风速

将各个方向的风速保存在列表中

```python
degs = np.arange(45,361,45)
degs
```

```python
tmp =  []
for deg in degs:
    #print milano[(milano['wind_deg']>(deg-46)) & (milano['wind_deg']<deg)]['wind_speed'].mean()
    tmp.append(milano[(milano['wind_deg']>(deg-46)) & (milano['wind_deg']<deg)]['wind_speed'].mean())
speeds = np.array(tmp)
print(speeds)
```

画出各个方向的风速

```python
show_rose(speeds,'milano')
```

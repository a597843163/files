分别使用线性回归，岭回归，KNN，以及ExtraTrees回归进行分析

```python
import numpy as np

from sklearn.linear_model import LinearRegression

# Knn回归
from sklearn.neighbors import  KNeighborsRegressor

from sklearn.linear_model import Ridge

# 对于数据的预测,更加的精准
from sklearn.ensemble import ExtraTreesRegressor

# 导入分析的数据
import sklearn.datasets as datasets

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
```

生成训练数据和预测数据

```python
# 这个是人脸信息
faces= datasets.fetch_olivetti_faces()
faces
```

```python
# data 和images是一样的数据,维度不一样
data = faces.data
images = faces.images
```

```python
display(data.shape,images.shape)
```

```python
plt.imshow(images[3],cmap='gray')
```

```python
# plt.figure(figsize=(2))
plt.imshow(data[0].reshape((64,64)),cmap='gray')
```

```python
target = faces.target
```

```python
# 人脸数据,分成上半部分和下半部分
# 300个完整的人脸  学习上半部分脸,给他目标下半部分脸
train = data[target<30]

# 100个完整的人脸,预测上半部分,预测的下半部分
test = data[target>=30]
```

```python
# 训练数据
X_train = train[:,:2048]
Y_train = train[:,2048:]
```

```python
face_upper = X_train[0].reshape((32,64))
```

```python
plt.imshow(face_upper,cmap='gray')
```

```python
# 预测数据
# x-test 会交给机器学习的算法,  ----->推导出来下半部分脸
x_test = test[:,:2048]
# 对照数据
y_test = test[:,2048:]
```

创建机器学习模型，以字典的形式包含四种模型

```python
# 创建四个模型字典
Estimator = {
    'Knn': KNeighborsRegressor(),
    'LinearRG': LinearRegression(),
    'Ridge': Ridge(),
    'ExtraTree': ExtraTreesRegressor()
}
```

使用模型训练数据并预测数据

```python
#保存预测出来的数据
y_predicts = dict()

for key,estimator in Estimator.items():
        
        # 第一步对模型进行训练
        estimator.fit(X_train,Y_train)
        
        # 使用模型进行数据的预测
        # 将上半部分脸传给模型,预测下半部分脸
        # 返回的就是下半部分脸
        y_ = estimator.predict(x_test)
        
        # 将数据保存到字典
        y_predicts[key] = y_
```

需要将数据进行可视化

```python
# 随机选5个数据来进行展示
face_ids = np.random.randint(0,100,size=5)

face_ids
```

```python
# 一张图片 2,2.4
# 5个数据.5行
# 4个算法,+true face  +半张脸 =6列
plt.figure(figsize=(6*2,5*2.4))

for i in range(5):
    # 绘制第一列 True Face
    # 
    if not i:
        axes = plt.subplot(5,6,i*6+1,title = 'True Face')
    else:
        axes = plt.subplot(5,6,i*6+1)
    # 绘制真脸数据
    axes.imshow(test[face_ids[i]].reshape(64,64),cmap='gray')
    
    # 绘制第二列的数据,半张脸
    if i :
        axes = plt.subplot(5,6,i*6+2)
    else:
        axes = plt.subplot(5,6,i*6+2,title = 'Half Face')
    # 绘制上半部分脸的数据
    axes.imshow(x_test[face_ids[i]].reshape(32,64),cmap='gray')
    # 绘制预测的数据
    # 预测的结果在字典中
    for j,key in enumerate(y_predicts):

        if i :
            axes = plt.subplot(5,6,i*6+(j+3))
        else:
            axes = plt.subplot(5,6,i*6+(j+3),title=key)
        
        # 绘制合成的人脸
        # 上半部分+下半部分
        # x_test   y_predicts
        face_upper = x_test[face_ids[i]]
        # 特定的算法,预测的数据
        y_ = y_predicts[key]
        
        face_lower = y_[face_ids[i]]
        comp_face = np.hstack((face_upper,face_lower)).reshape(64,64)
        
        axes.imshow(comp_face,cmap='gray')
    
```

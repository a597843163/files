```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
# 读取数据  sep可以把文件里面的制表符识别
cancer = pd.read_csv('cancer.csv',sep='\t')
```

```python
cancer.shape
```

```python
cancer.head()
```

```python
# 获取训练数据
target = cancer[['Diagnosis']]
target
```

```python
cancer.drop(['ID','Diagnosis'],axis=1,inplace=True)
```

```python
# 分割数据   将数据进行分割,取0.1部分数据来测试,其他都是训练数据
X_train,x_test,Y_train,y_test = train_test_split(cancer,target,test_size=0.1)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,Y_train['Diagnosis'].values)
```

```python
knn.score(X_train,Y_train['Diagnosis'].values)
```

```python
# 训练
y_ = knn.predict(x_test)
```

```python
y_
```

```python
y_test['Diagnosis'].values
```

```python
# 交叉表   index,columns
pd.crosstab(y_,y_test['Diagnosis'].values,rownames=['预测'],colnames=['真实'],margins=True)
```

```python
knn.score(x_test,y_test)
```
